```python
# Data Visualization
def plot_data_distributions(data):
    data.hist(figsize=(12, 10))
    plt.suptitle('Data Distribution for Numerical Features')
    plt.savefig('data_distribution.png')
    plt.show()

def plot_categorical_distributions(data, column):
    plt.figure(figsize=(8, 6))
    data[column].value_counts().plot(kind='bar')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.title(f'Distribution of {column}')
    plt.savefig(f'{column}_distribution.png')
    plt.show()

def plot_correlation_matrix(data):
    plt.figure(figsize=(10, 10))
    sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.savefig('correlation_matrix.png')
    plt.show()

# Use the original data for visualizations
data = pd.read_csv(file_path)  # Load original data again
plot_data_distributions(data)
plot_categorical_distributions(data, 'deposit')
plot_correlation_matrix(data)
```
